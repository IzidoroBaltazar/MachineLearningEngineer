```python
import quandl
import pandas as pd
from time import sleep
import matplotlib.pyplot as plt
from datetime import date
import holidays

start_date = '1996-1-1'
end_date = '2018-12-22'

cftc_gold = "088691"
cftc_corn = "002602"
cftc_coffee = "083731"

with open('apitoken', 'r') as file:
    apitoken = file.read().strip()

def get_frames(name, token):
    all_months = ['F', 'G', 'H', 'J', 'K', 'M', 'N', 'Q', 'U', 'V', 'X', 'Z']
    months = ['F', 'H', 'K', 'N', 'U', 'Z']
    months = all_months
    frames = []
    srf = name
    for year in range(1996,2019):
        for month in months:
            ticker = srf + month + str(year)
            print(ticker)
            try:
                frames.append(quandl.get(ticker, authtoken=token))
            except Exception as e:
                print(e)
                sleep(1)
    return frames
```

```python
%%capture
gold_frames = get_frames('SRF/CME_GC', apitoken)
corn_frames = get_frames('SRF/CME_C', apitoken)
coffee_frames = get_frames('SRF/ICE_KC', apitoken)
```

```python
keys = []
for df in (gold_frames + corn_frames + coffee_frames):
    for key in df.keys():
        if key not in keys:
            print(key)
            keys.append(key)
```

```python
def put_together_by_volume(frames):
    global start_date, end_date
    df = pd.concat(frames)
    idx = df.groupby(['Date'])['Volume'].transform(max) == df['Volume']
    ans_df = df[idx]
    idx = ans_df.groupby(['Date'])['Prev. Day Open Interest'].transform(max) == ans_df['Prev. Day Open Interest']
    ans_df = ans_df[idx]
    ans_df = ans_df.loc[(ans_df.index > start_date) & (ans_df.index < end_date)]
    return ans_df
```

```python
import requests
import io
import zipfile


def download_extract_zip(url):
    """
    Download a ZIP file and extract its contents in memory
    yields (filename, file-like object) pairs
    """
    response = requests.get(url)
    with zipfile.ZipFile(io.BytesIO(response.content)) as thezip:
        for zipinfo in thezip.infolist():
            with thezip.open(zipinfo) as thefile:
                yield zipinfo.filename, thefile
                

def download_save_cot(year):
    yr = str(year)
    url = "https://www.cftc.gov/files/dea/history/deacot" + yr + ".zip"
    report = download_extract_zip(url)
    file_name = "cot_reports/cot_" + yr + ".csv"
    with open(file_name, "wb") as file:
        for record in report:
            name, fl = record
            file.write(fl.read())
```

```python
for year in range(1996, 2019):
    download_save_cot(year)
```
